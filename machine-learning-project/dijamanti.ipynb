```python
plt.figure(figsize = (16,5))
sns.heatmap(corr, annot=True)
```

Varijable 'depth' i 'table' imaju najmanju korelaciju sa svim ostalim varijablama. Prvi poriv je izbaciti te parametre iz analize. No, niski koeficijent korelacije ne znači da te značajke neće doprinijeti kvaliteti modela.

Inače se najviše problema stvara kod varijabli koje su visoko korelirane ili čak kolinearne. Nasuprot naivnoj pretpostavci, visoka međusobna koreliranost značajki za predviđanje ne poboljšava moć modela da predvidi nezavisnu varijablu. Čak i ako se izbace neke od međusobno koreliranih značajki, i dalje postoji opasnost od pojave da tada prijašnje nisko korelirane značajke postanu visoko korelirane.

Izbor varijabli za regresijsko predviđanje je kompliciran posao i često se razlikuje od jednog skupa podataka do drugog (zbog razdiobe podataka ili nekih drugih utjecaja).

Ovdje se nije biralo značajke nego se uzelo sve. To ne garantira najbolju moć modela, iako je intuitivno da više značajki (a time i više informacija) mora rezultirati boljim predviđanjem. Postoje metode za izbor značajki (npr. na temelju korelacije, dobitka informacija - "information gain", itd.).

Umjesto ugrađenih metoda isprobalo se nekoliko kombinacija značajki, ali sve kombinacije su dale jako slične rezultate pa se na kraju ta analiza nije uključila.

```python
for col in numerical.columns:
    plt.title("Boxplot for {0}".format(col))
    plt.boxplot(numerical[col])
    plt.show()
```

Na kutijastim dijagramima vidimo obilje outliera, što i ne čudi previše jer zbirka dijamanata ovog uzorka sadrži razno razne vrste dijamanata a u skladu s tim postoji i velika varijacija u cijeni.

Iako je prva ideja što pada na pamet izbaciti sve outliere, to vjerojatno nije dobra ideja za ovaj skup podataka jer outlieri sačinjavanju veliki dio podataka. Vrlo je lako moguće da je ova pretpostavka netočna, ali s obzirom na kasnije dobivene dobre rezultate, ostavljeno je ovako.

## Provođenje regresije

Ideja: predviđanje parametra "price" pomoću parametara "carat",  "cut",  "color",  "clarity",  "depth",  "table", "x", "y", "z". Prvi korak: podjela podataka na skup za treniranje i skup za testiranje. Ugrađena sklearn funkcija u pozadini ima logiku nasumičnosti, tj. neće jednostavno pocijepati skup na dva dijela, nego ga i permutira.

```python
from sklearn.model_selection import train_test_split
train_X, test_X = train_test_split(data)
train_y = train_X['price']
test_y = test_X['price']
```

```python
train_X.shape, test_X.shape
```

Ispuštanje parametara koje nećemo koristiti za predviđanje cijene, uključujući ispuštanje same cijene iz podataka koji će sačinjavati matricu značajki (feature matrix).

```python
train_X = train_X.drop(['Unnamed: 0', 'price'], axis=1)
test_X = test_X.drop(['Unnamed: 0', 'price'], axis=1)
train_X.shape, test_X.shape
```

Kodiranje kategoričkih varijabli numeričkim vrijednostima. Kako su kategoričke varijable rangirane, koristi se rastući niz prirodnih brojeva.

Postoji sklearn ugrađena funkcija za pretvorbu kategoričkih varijabli u numerički "ekvivalent", ali to se ovdje nije koristilo jer su i s ovim naivnim pristupom i dalje dobiveni dobri rezultati.

```python
cut = lambda x : 1 if x=='Fair' else ( 2 if x=='Good' else ( 3 if x=='Very Good' else ( 4 if x=='Premium' else (5))))
train_X['cut'] = train_X['cut'].apply(cut)
test_X['cut'] = test_X['cut'].apply(cut)

color = lambda x : 1 if x=='D' else ( 2 if x=='E' else ( 3 if x=='F' else ( 4 if x=='G' else ( 5 if x=='H' else ( 6 if x=='I' else (7))))))
train_X['color'] = train_X['color'].apply(color)
test_X['color'] = test_X['color'].apply(color)

clarity = lambda x : 1 if x=='I1' else ( 2 if x=='SI2' else ( 3 if x=='SI1' else ( 4 if x=='VS2' else ( 5 if x=='VS1' else ( 6 if x=='VVS2' else ( 7 if x=='VVS1' else (8)))))))
train_X['clarity'] = train_X['clarity'].apply(clarity)
test_X['clarity'] = test_X['clarity'].apply(clarity)

train_X, test_X
```

Pretvaranje Pandas dataframe objekata u Numpy array objekte zbog kasnije korištenih sklearn algoritama koji očekuju ulaz tog tipa.

```python
train_X = train_X.to_numpy()
test_X = test_X.to_numpy()
train_X
```

Prije samog pokušaja predviđanja, potrebno je standardizirati podatke. Trenutno nisu balansirani, a regresor koji će se koristiti pretpostavlja da su numeričke vrijednosti koje mu se proslijede normalno distribuirane.

Standardizacijom se podaci centriraju i skaliraju.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)
train_X
```

### Linear support vector regresor

Vrsta regresora koji će se korisititi je linear support vector regressor. To je vrsta support vector machine koja za izlaz daje predviđene numeričke vrijednosti umjesto klase, čime od klasifikatora postaje regresor.

```python
from sklearn.svm import LinearSVR
```

Regresoru je potrebno samo proslijediti pripremljene predprocesuirane podatke za predviđanje, i to u obliku feature matrice i točnih meta koje je iz nje potrebno predvidjeti. Ovo je vrsta nadziranog (supervised) učenja, jer osim materijala korištenog za predviđanje izlaza, algoritam dobiva i očekivani izlaz.
Dodatne postavke koje se mogu namjestiti se vide u posljednjoj naredbi u vidu hiperparametara (metavarijabli za algoritam).
- c = regularizacijski parametar
- dual = bool vrijednost kojom se namješta hoće li algoritam rješavati dualni optimizacijski problem
- epsilon = varijabla epsilon iz epsilon-insensitive funkcije gubitka (loss function)
- fit_intercept = bool vrijednost koja određuje hoće li se računati intercept (odsjek). ako ne, onda se pretpostavlja da su podaci centrirani
- intercept_scaling = parametar koji određuje jakost regularizacije provođenom nad "sintetički stvorenim" značajkama koje se dodaju originalnima ako je fit_intercept == True
- loss = vrsta funkcije gubitka koja se primjenjuje. ovdje se koristi epsilon-insensitive funkcija gubitka (koja odgovara L1 normi), a mogla se i squared epsilon-insensitive funkcija gubitka (koja odgovara L2 normi)
- max_iter = najveći dozvoljeni broj iteracija prije stajanja algoritma
- random_state = kontrola generatora slučajnih brojeva. namješteno na fiksni cijeli broj kako bi se za više poziva funkcije mogli reproducirati isti rezultati računa
- tol = tolerancija za kriterij stajanja
- verbose = kontrola izlaza (verbose ili ne)

```python
regr = LinearSVR(random_state=0, tol=1e-5)
regr.fit(train_X, train_y)
```

Nakon "fittanja" (učenja) modela, mogu se ispisati neki njegovi dobiveni parametri. Npr. koeficijenti hiperravnine i odsječak.

```python
print(regr.coef_)
print(regr.intercept_)
```

Sada je moguće na testnim podacima provjeriti valjanost modela. Pogledajmo kakve predviđene vrijednosti izlaza se dobivaju iz predviđanja na testnom skupu značajki.

```python
p = regr.predict(test_X)
print(p)
```

Gomila brojeva koja nam ništa puno ne znači. Zbog toga provjerimo uspjeh modela ugrađenom metodom.

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

greska = mean_squared_error(test_y, p)
print(sqrt(greska))
```

Srednja kvadratna pogreška je jedna od osnovnih mjera valjanosti regresijskog modela. Da bi imali smisleniji dojam o uspješnosti modela, koristi se ugrađena sklearn funkcija score() koja ovisno o modelu primjenjuje prikladnu metriku i vraća informaciju o uspješnosti predviđanja u obliku postotka.

```python
regr.score(test_X, test_y)
```

85% nije loš rezultat. Vjerojatno ga je moguće poboljšati; namještanjem hiperparametara SVR, dodavanjem još podataka, ili u krajnju ruku korištenjem drugog regresora. Ispitajmo zadnju hipotezu.

Proučimo malo parametre stabla.

- max_depth = maksimalna dubina stabla
- criterion = mjera kvalitete grananja. default je "mse" (mean squared error).
- splitter = kriterij grananja. može biti i "random"
- min_samples_split = minimalni broj uzoraka koji mora biti prisutan u unutarnjem čvoru da bi nastupilo grananje, default 2
- min_samples_leaf = minimalni broj uzoraka koji mora biti prisutan u čvoru da bi bio list
- random_state = sjeme (seed) pseudo random number generatora. default je None što odgovara numpy random number generator objektu
- max_leaf_nodes = maksimalni broj listova
- min_impurity_decrease = pomoć kod grananja, grananje će se dogoditi ako vrijednost nečistoće (impurity) pada za više od ove zadane vrijednosti nakon podjele
- min_impurity_split = granica za rano stajanje u rastu stabla
- ccp_alpha = parametar kompleksnosti korišten za Minimal cost-complexity pruning (rezanje na temelju najmanjeg troška i kompleksnosti)

```python
clf = tree.DecisionTreeRegressor(max_depth=3, min_samples_split=5)
clf = clf.fit(train_X, train_y)

predicted = clf.predict(test_X)
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, feature_names=['carat',  'cut',  'color',  'clarity', 'depth', 'table', 'x', 'y', 'z'], filled=True)
```

Promotrimo score svakog od stabala.

```python
clf2.score(test_X, test_y), clf5.score(test_X, test_y), clf.score(test_X, test_y)
```

Vidi se da je najbolje bilo stablo s najvećom dubinom (5), a najgore ono s najmanjom dubinom (2).

### Partial least squares regresor

Iduća metoda regresije koja će se probati je metoda parcijalnih najmanjih kvadrata. Ovaj algoritam je nadogradnja na PCR (principal component regression) koji je regresijski oblik modela za smanjenje dimenzionalnosti podataka, PCA (principal component analysis).

Osnovna razlika između PLS i PCR je ta što PLS inkorporira vrijednosti izlaza y u proces smanjenja dimnezionalnosti, čime se (teoretski) smanjenjem prostora djelovanja ne gube potencijalno bitne značajke za predviđanje koje bi PCR izgubio jer imaju malu varijancu.

```python
from sklearn.cross_decomposition import PLSRegression

pls2 = PLSRegression(n_components=2)
pls2.fit_transform(train_X, train_y)
```

```python
predicted_pls2 = pls2.predict(test_X)
predicted_pls2
```

```python
pls2.score(test_X, test_y)
```

Nešto bolji rezultat nego što je dao support vector regressor i neka od stabala, ali i dalje nije nadmašeno najdublje stablo. To je moguće i zbog overfittinga stabla. Tome bi se moglo doskočiti raznim tehnikama, od kojih će se ovdje probati prunanje (rezidba) stabla.
